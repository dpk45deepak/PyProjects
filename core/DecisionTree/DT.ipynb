## Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.keys()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
y = pd.DataFrame(iris.target, columns=['target'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
#  apply decision tree classifier model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
# visualize the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(15,10))
plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

```python
# how to prevent the overfitting of the decision tree model
# apply decision tree classifier model
# because the default value of max_depth is None, the decision tree model will be overfitted to the training data. To prevent the overfitting, we can set the max_depth parameter to a smaller value.
# the max_depth parameter is the maximum depth of the decision tree. The default value is None, which means that the decision tree will be expanded until all leaves are pure or until all leaves contain less than min_samples_split samples. By setting the max_depth parameter to a smaller value, we can prevent the overfitting of the decision tree model.
# the max_depth parameter is a hyperparameter that we can tune to find the best value for our model. We can use cross-validation to find the best value for the max_depth parameter.
model2 = DecisionTreeClassifier(max_depth=2)
model2.fit(X_train, y_train)
```

```python
# visualize the decision tree
plt.figure(figsize=(15,10))
plot_tree(model2, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

```python
# print the confusion matrix for both models
from sklearn.metrics import confusion_matrix, classification_report
y_pred = model.predict(X_test)
y_pred2 = model2.predict(X_test)
print("Confusion Matrix for model 1:")
print(confusion_matrix(y_test, y_pred))
print("Confusion Matrix for model 2:")
print(confusion_matrix(y_test, y_pred2))
print("Classification Report for model 1:")
print(classification_report(y_test, y_pred))
print("Classification Report for model 2:")
print(classification_report(y_test, y_pred2))
```

```python
#  dump the decision tree model to a file
import joblib
joblib.dump(model, '../../models/DecisionTree/decision_tree_model.joblib')
```

## hyper parameter tuning and decision tree classifier pruning

```python
params = {
    'max_depth': [2, 3, 4, 5, 6],
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=params, cv=5)
grid_search.fit(X_train, y_train)
```

```python
# print the best parameters and the best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
# print the best parameters and the best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

# the best parameters are {'criterion': 'gini', 'max_depth': 2, 'max_features': 'auto', 'splitter': 'best'} and the best score is 0.9666666666666668.
# print the confusion matrix and classification report for the best model
# print accuracy score for the best model
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

y_pred = grid_search.predict(X_test)
print("Confusion Matrix for the best model:")
print(confusion_matrix(y_test, y_pred))
print("Classification Report for the best model:")
print(classification_report(y_test, y_pred))
print("Accuracy Score for the best model:")
print(accuracy_score(y_test, y_pred))
```
